```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from datetime import datetime as dt
from pprint import pprint
from ipywidgets import widgets
```

```python
data = pd.read_excel("ferry_schedule.xlsx")
```

```python
data
```

```python
df = data.melt(
    id_vars=["Stop", "Direction"],
    value_name="Time",
    var_name="Order"
)
```

```python
df["Time"] = np.where(df["Time"].str.contains("[0-9]+", regex=True), df["Time"], None)
```

```python
df = df.loc[~pd.isnull(df["Time"])]
```

```python
df["Time"] = pd.to_datetime(df["Time"].str.lstrip().str.rstrip(), format="%I:%M %p").apply(lambda x: x.replace(year=1970, month=1, day=1))
```

```python
df = df.reset_index().drop("index", axis=1)
```

```python
df
```

```python
stops = df[["Stop", "Direction", "Order"]].drop_duplicates().values.tolist()
# data is a dictionary with two keys. Each key's value is a list
lines = {"line_x": [], "line_y": [], "direction": [], "group": []}

# Loop over unique matchups
for i, s in enumerate(stops):
    val = df.loc[
        (df["Stop"] == s[0]) 
        & (df["Direction"] == s[1]) 
        & (df["Order"] == s[2])
    ]["Time"].values

    if i > 0:
        prev_s = stops[i-1]
        
        if s[2] > prev_s[2] or s[1] != prev_s[1]:
            lines["line_x"].extend([None])
            lines["line_y"].extend([None])
            lines["direction"].extend([None])
            lines["group"].extend([s[2]])
        
        elif s[2] > prev_s[2]:
            lines["group"].extend([None])
        else:
            pass
            
    if len(val) > 0:
        lines["line_x"].extend(
            [
                val[0],
            ]
        )
        lines["line_y"].extend([s[0]])
        lines["direction"].extend([s[1]])
        lines["group"].extend([s[2]])
    
```

```python
def update_trace(trace, points, selector):
    if len(points.point_inds) == 1:
        i = points.trace_index
        for x in range(0, len(fig.data)):
            fig.data[x]['line']['color'] = 'grey'
            fig.data[x]['opacity'] = 0.7
#             fig.data[x]['line']['width'] = default_linewidth
        pprint([
            (el[0], f"{el[1].hour}:{el[1].minute:02d}", el[2][0]) 
            for el in zip(fig.data[i]["y"], fig.data[i]["x"], fig.data[i]["customdata"])
            if el[0] is not None
        ])
        fig.data[i]['line']['color'] = 'red'
        fig.data[i]['opacity'] = 1
#         fig.data[i]['line']['width'] = highlighted_linewidth
```

```python
fig = go.FigureWidget()
for i in tmp_df["group"].unique():
    lines = tmp_df.loc[tmp_df["group"] == i]
    fig.add_trace(
        go.Scatter(
            # Line between dots
            x=pd.to_datetime(lines["line_x"], format=""),
            y=lines["line_y"],
            customdata=np.stack((lines["direction"], lines["group"]), axis=-1),
            mode="markers+lines",
            showlegend=False,
            fillcolor="black",
            opacity=0.7,
            marker=dict(
                symbol="arrow",
                color="black",
                size=8,
                # This makes the arrow point the correct direction
                angleref="previous",
                # This shortens the length of the arrow so it points to the dot
                #   and not into the dot
                standoff=0
            ),
            hovertemplate="Stop: %{y}<br>"
                          "Time: %{x}<br>"
                          "Direction: %{customdata[0]}<br>"
                          "<extra></extra>"
        ),
    )
    
fig.update_layout(
        margin=dict(l=0, r=0, t=20, b=20),
        xaxis=dict(
            range=["1970-01-01 05:30:00", "1970-01-01 23:00:00"],
            tickformat = '%H:%M',
            dtick=900000,
    #         tickformatstops=[
    #             dict(dtickrange=["05:30", "23:00"], value="%I:%M")
    #         ],
            type="date",
            showgrid=True,
            gridcolor="lightgray",
            automargin=True
        ),
        yaxis=dict(
            type="category",
            categoryorder='array',
            categoryarray=["Hingham", "Hull",
                           "Logan", "Rowes Wharf", "Long Wharf"],
            autorange="reversed",
            showgrid=True,
            gridcolor="lightgray",
        ),
        plot_bgcolor="white",
        hovermode="closest",
    height=1000
    )

for x in range(0, len(fig.data)):
    fig.data[x].on_hover(update_trace)
```
